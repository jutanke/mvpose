```python
%matplotlib inline
import matplotlib.pyplot as plt

# kth different sigmas

sigma = [0.5, 1, 1.5, 1.8, 2, 2.3, 2.5, 2.8, 3, 5]
values = [.93, .957, .9667, .96904, .9690, .9684, .96728, .96495, .9626, .9345]
assert len(values) == len(sigma)


fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ax.plot(sigma, values)

plt.show()
```
