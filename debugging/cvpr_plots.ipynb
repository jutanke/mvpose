```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
plt.rcParams.update({'font.size': 16})

# kth different sigmas

sigma = [0.5, 1, 1.5, 1.8, 2, 2.3, 2.5, 2.8, 3, 5, 6]
values_kth = [.93, .957, .9667, .96904, .9690, .9684, .96728, .96495, .9626, .9345, .9077]


print(len(values_kth))

sigma_camp = [0.5, 1, 1.5, 1.8, 2, 2.3, 2.5, 2.8, 3, 3.2, 3.5, 3.8, 4, 4.2, 4.5, 5, 6]
values_camp = [.8863,.89154, .8914, .8955, .89739, .8985, .897909, .90106, .90299, .9047, .90522,
              .906847, .9057, .90693, .9078, .9057, .901]

sigma_shelf = [0.5, 1, 1.5, 1.7, 1.8, 2, 3, 5, 6]
values_shelf = [.9247, .9327, .933165, .933056, .9308, .92768, .9288, .92201,.9196]

assert len(values_kth) == len(sigma)
assert len(values_camp) == len(sigma_camp)
assert len(values_shelf) == len(sigma_shelf)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
ax.plot(sigma, values_kth, label='KTH Football II', linewidth=3)
ax.plot(sigma_camp, values_camp, label='Campus', linewidth=3)
ax.plot(sigma_shelf, values_shelf, label='Shelf', linewidth=3)
ax.set_xlabel(r"$\sigma$", fontsize=30)
ax.set_ylabel("PCP", fontsize=20)

plt.legend(fontsize=16)

# plt.show()
plt.savefig('sigma.eps', format='eps', dpi=1000)
```
